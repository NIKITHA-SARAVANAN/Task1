## Bins Parameter

```python
x=[1,12,22,21,20,21,31,2,32,40,14,33,50,44,45,46,45,44,46,47,43,42,41,40,43,45,43,44,61,13,58,70,12,60,10,71,11,72,
   18,85,90,96,99,25,35,23,33,34,13,28,25,13,34,45,55,56,64,73,80,100]
y=[1,11,21,31,41,51,61,71,81,91,101]
plt.hist(x,y,ec="black")
plt.xlabel("age")
plt.ylabel("age count")
plt.title("Population age count")
plt.xticks(y)
plt.show()
```

##  Using range

```python
x=[1,12,22,21,20,21,31,2,32,40,14,33,50,44,45,46,45,44,46,47,43,42,41,40,43,45,43,44,61,13,58,70,12,60,10,71,11,72,
   18,85,90,96,99,25,35,23,33,34,13,28,25,13,34,45,55,56,64,73,80,100]
plt.hist(x,20,(0,100),ec="red")
plt.xlabel("age")
plt.ylabel("age count")
plt.title("Population age count")
plt.show()
```

## Implementing weight paramater

```python
x=[1,12,22,21,20,21]
y=[1,11,21,31]
plt.hist(x,y,ec="red",weights=[2,1,2,1,2,2])
plt.xlabel("age")
plt.ylabel("age count")
plt.title("Population age count")
plt.show()
```
